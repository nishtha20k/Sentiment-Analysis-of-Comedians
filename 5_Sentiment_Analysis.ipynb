```python
import numpy as np
import math

def split_text(text, n=20):
    length = len(text)
    size = math.floor(length / n)
    start = np.arange(0, length, size)

    split_list = []
    for piece in range(n):
        split_list.append(text[start[piece]:start[piece]+size])
    return split_list
```

```python
import pandas as pd
from textblob import TextBlob

data_data = pd.read_pickle('corpus.pkl')

# Defined a threshold ( 0.1 and -0.1) to classify a sentence as emotional or neutral based on its polarity score.
# Sentences with polarity scores greater than 0.1 are considered emotional, those with scores less than -0.1 are considered emotional in the opposite direction, and the rest are considered neutral.
def calculate_sentiment(text):
    blob = TextBlob(text)
    emotional_score = 0
    neutral_score = 0
    for sentence in blob.sentences:
        if sentence.sentiment.polarity > 0.1:
            emotional_score += 1
        elif sentence.sentiment.polarity < -0.1:
            emotional_score -= 1
        else:
            neutral_score += 1
    return emotional_score, neutral_score

sentiment_results = []
for comedian, transcript in zip(data_data.index, data_data['transcript']):
    emotional_score, neutral_score = calculate_sentiment(transcript)
    sentiment_results.append({'comedian': comedian, 'emotional_score': emotional_score, 'neutral_score': neutral_score})

sentiment_df = pd.DataFrame(sentiment_results)
print(sentiment_df)
```

```python
# Plotting emotional score
plt.figure(figsize=(10, 6))
plt.bar(sentiment_df['comedian'], sentiment_df['emotional_score'], color='blue')
plt.xlabel('Comedian')
plt.ylabel('Emotional Score')
plt.title('Emotional Score for Each Comedian')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Plotting neutral score
plt.figure(figsize=(10, 6))
plt.bar(sentiment_df['comedian'], sentiment_df['neutral_score'], color='green')
plt.xlabel('Comedian')
plt.ylabel('Neutral Score')
plt.title('Neutral Score for Each Comedian')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
