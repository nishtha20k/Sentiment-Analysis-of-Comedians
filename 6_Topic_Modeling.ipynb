**Attempt-1**

```python
import pandas as pd
import pickle

data = pd.read_pickle('dtm_stop.pkl')
data
```

```python
from gensim import matutils, models
import scipy.sparse
```

```python
tdm = data.transpose()
tdm.head()
```

```python
sparse_counts = scipy.sparse.csr_matrix(tdm) # converts a term-document matrix (tdm) into a sparse matrix using Compressed Sparse Row format (csr_matrix)
corpus = matutils.Sparse2Corpus(sparse_counts) # converts the sparse matrix into a Gensim corpus object
```

```python
cv = pickle.load(open("cv_stop.pkl", "rb")) # Load pickled file in binary read mode
id2word = dict((v, k) for k, v in cv.vocabulary_.items()) # The vocabulary is typically a dictionary where keys are words and values are their corresponding integer IDs (key-value pairs).
```

```python
lda = models.LdaModel(corpus=corpus, id2word=id2word, num_topics=2, passes=50)
lda.print_topics()
```

```python
lda = models.LdaModel(corpus=corpus, id2word=id2word, num_topics=3, passes=50)
lda.print_topics()
```

```python
lda = models.LdaModel(corpus=corpus, id2word=id2word, num_topics=4, passes=50)
lda.print_topics()
```

* Despite increasing the number of topics and running multiple iterations, there isn't much variation observed in the topics generated.
* Certain words such as audience, laughing, applauding and others are repeated across different topics, indicating that these terms might not be effectively distinguishing between topics.

```python
from nltk import word_tokenize, pos_tag

def nouns(text):
    is_noun = lambda pos: pos[:2] == 'NN'
    tokenized = word_tokenize(text)
    all_nouns = [word for (word, pos) in pos_tag(tokenized) if is_noun(pos)]
    return ' '.join(all_nouns)
```

```python
data_clean = pd.read_pickle('data_clean.pkl')
data_clean
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
data_nouns = pd.DataFrame(data_clean.transcript.apply(nouns))
data_nouns
```

```python
def nouns_adj(text):
    is_noun_adj = lambda pos: pos[:2] == 'NN' or pos[:2] == 'JJ'
    tokenized = word_tokenize(text)
    nouns_adj = [word for (word, pos) in pos_tag(tokenized) if is_noun_adj(pos)]
    return ' '.join(nouns_adj)
```

```python
data_nouns_adj = pd.DataFrame(data_clean.transcript.apply(nouns_adj))
data_nouns_adj
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

add_stop_words = ['like', 'im', 'know', 'just', 'dont', 'thats', 'right', 'people','youre', 'got', 'gonna', 'time', 'think', 'yeah', 'said']
stop_words = list(text.ENGLISH_STOP_WORDS.union(add_stop_words))

cvna = CountVectorizer(stop_words=stop_words, max_df=.8)
data_cvna = cvna.fit_transform(data_nouns_adj.transcript)
data_dtmna = pd.DataFrame(data_cvna.toarray(), columns=cvna.get_feature_names_out())
data_dtmna.index = data_nouns_adj.index
data_dtmna
```

```python
corpusna = matutils.Sparse2Corpus(scipy.sparse.csr_matrix(data_dtmna.transpose()))
id2wordna = dict((v, k) for k, v in cvna.vocabulary_.items())
```

```python
ldana = models.LdaModel(corpus=corpusna, num_topics=2, id2word=id2wordna, passes=50)
ldana.print_topics()
```

```python
ldana = models.LdaModel(corpus=corpusna, num_topics=3, id2word=id2wordna, passes=50)
ldana.print_topics()
```

```python
ldana = models.LdaModel(corpus=corpusna, num_topics=4, id2word=id2wordna, passes=50)
ldana.print_topics()
```

```python
ldana = models.LdaModel(corpus=corpusna, num_topics=4, id2word=id2wordna, passes=80)
ldana.print_topics()
```

* Topic 0: Personal Relationships (e.g., "mom," "dad")
* Topic 1: Emotional Expressions and Contextual Language (e.g., "fuck," "pussy")
* Topic 2: Social Identity and Interaction (e.g., "black," "gay," "dude")
* Topic 3: Miscellaneous or Unclear (e.g., "poor," "pubes," "music")

```python
corpus_transformed = ldana[corpusna]
topics = [topics_tuple[0] for topics_tuple in corpus_transformed]
topic_document_pairs = list(zip(topics, data_dtmna.index))
topic_document_df = pd.DataFrame(topic_document_pairs, columns=['Topic', 'Document Index'])
print(topic_document_df)
```
